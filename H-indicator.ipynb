```python
import pandas as pd
import numpy as np
from dzh_utils import match_i_Huskies_passing_table
def sub_pitch_passing_table(matchID, teamID):
    df = pd.read_csv('./2020_Problem_D_DATA/passingevents.csv')
    df = df[df['MatchID'] == matchID]
    df = df[df['TeamID'] == teamID]
    df = df.reset_index(drop=True)
    table = {(x,y):[0,0]for x in range(101) for y in range(101)}
    for i in range(len(df)):
        sub_pitch_x_o, sub_pitch_y_o = cal_sub_pitch_pos(df['EventOrigin_x'][i], df['EventOrigin_y'][i])
        sub_pitch_x_d, sub_pitch_y_d = cal_sub_pitch_pos(df['EventDestination_x'][i], df['EventDestination_y'][i])
        table[(sub_pitch_x_o, sub_pitch_y_o)][0] += 1
        table[(sub_pitch_x_d, sub_pitch_y_d)][1] += 1    
    return table

def cal_sub_pitch_pos(x, y):
    return int(x//1), int(y//1)

def H_indicator(matchID, teamID):
    df = pd.read_csv('./2020_Problem_D_DATA/passingevents.csv')
    df = df[df['MatchID'] == matchID]
    df = df[df['TeamID'] == teamID]
    w = len(df) * 2
    match_id_Huskies_passing_table = match_i_Huskies_passing_table(matchID)
    data = np.array([value[0]+value[1] for key, value in match_id_Huskies_passing_table.items()])
    mu_p = data.mean()
    sigma_p = np.sqrt(data.var())
    data = np.array([value[0]+value[1] for key, value in sub_pitch_passing_table(matchID, teamID).items()])
    mu_z = data.mean()
    sigma_z = np.sqrt(data.var())
    h = 5/(1/w+1/mu_p+1/sigma_p+1/mu_z+1/sigma_z)
    return h
```
